```python
#to check the impact of parents education on student 
gb = df.groupby("ParentEduc").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)                                  
```

```python
sns.heatmap(gb)
plt.show()
```

```python

plt.figure(figsize=(4,4))
sns.heatmap(gb, annot = True)
plt.title("Impact of Parents Education")
plt.show()
```

```python
gb = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean', "ReadingScore":'mean', "WritingScore":'mean'})
print(gb)    
```

```python
plt.figure(figsize=(4,4))
sns.heatmap(gb, cmap="viridis", annot=True)
plt.title("Impact of Parents marital status")
plt.show()
```
